```python
cerebra = CerebrA()
cerebra.label_details
import matplotlib.pyplot as plt
import random
import matplotlib.colors
ax = plt.figure().add_subplot(projection='3d')


region_ids = [3,11,56]
colors = [(0.2,0.2,0.6),(0.5,0.2,0.6),(0.2,0.3,0.1)]


# rid = 11
# region_ids = [rid, rid + 51]
# colors = [(0.2,0.2,0.6),(0.5,0.2,0.6)]

# region_ids = [11]
# colors = [(0.2,0.2,0.6)]

name = ""

for region_id, color in zip(region_ids, colors):
    points = cerebra.get_points_from_region_id(region_id).T

    xs = points[0]
    ys = points[1]
    zs = points[2] 
    cs = [color for i in range(len(xs))]

    ax.scatter(xs,ys,zs,c=cs, alpha=1)
    name += " " + cerebra.get_region_name_from_region_id(region_id) + ", "
    
    x,y,z = cerebra.find_region_centroid_from_id(region_id)
    
    ax.scatter(x,y,z,c='red', alpha=1)
    
ax.set_title(name)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")


ax.set_xlim([0,256])
ax.set_ylim([0,256])
ax.set_zlim([0,256])
```
